```python
# Converting Date column into datetime format

temperature_df['Date']=pd.to_datetime(temperature_df['Date'])
```

```python
# Removing Leap Day

temperature_df=temperature_df[~((temperature_df['Date'].dt.month==2)& (temperature_df['Date'].dt.day==29))]
```

```python
temp_2005_2014=temperature_df[(temperature_df['Date'].dt.year>=2005) & (temperature_df['Date'].dt.year<=2014)]
temp_2015=temperature_df[temperature_df['Date'].dt.year==2015]
```

```python
# Extract Day of the year

temp_2005_2014["DayOfYear"]=temp_2005_2014["Date"].dt.strftime("%m-%d")
temp_2015["DayOfYear"]=temp_2015["Date"].dt.strftime("%m-%d")
```

```python
# Record Highs & Low

record_highs=temp_2005_2014[temp_2005_2014["Element"]=="TMAX"].groupby("DayOfYear")["Data_Value"].max()
record_lows=temp_2005_2014[temp_2005_2014["Element"]=="TMIN"].groupby("DayOfYear")["Data_Value"].min()
```

```python
record_highs.head()
```

```python
# 2015 records breaking prvs high/lows

high_2015=temp_2015[temp_2015["Element"]=="TMAX"].groupby("DayOfYear")["Data_Value"].max()
low_2015=temp_2015[temp_2015["Element"]=="TMIN"].groupby("DayOfYear")["Data_Value"].min()
```

```python
high_break=high_2015[high_2015>record_highs]
low_break=low_2015[low_2015<record_lows]
```

```python
# Plotting

# Convert index to numeric day of year

# So by this we have got the lows and highest from 2005-2014 vs 2015 in this plot
high_break.index = pd.to_datetime(high_break.index, format="%m-%d").day_of_year
low_break.index = pd.to_datetime(low_break.index, format="%m-%d").day_of_year

plt.figure(figsize=(12,6))
plt.fill_between(record_highs.index, record_highs.values, record_lows.values, color='gray', alpha=0.3, label="2005-2014 range")
plt.scatter(high_break.index, high_break.values, color='red', label="2015 Record Highs", s=50, zorder=3)
plt.scatter(low_break.index, low_break.values, color='blue', label="2015 Record Lows", s=50, zorder=3)
plt.xlabel("Day of the Year")
plt.ylabel("Temperature (tenths of C)")
plt.title("Record High and Low Temperature (2005-2014) with 2015 Breaks")
plt.xticks(ticks=range(0, 366, 30), labels=pd.date_range("2023-01-01", periods=13, freq="M").strftime("%b"))
plt.legend()
plt.grid()
plt.show()
```

```python
# Map station locations
stations = bin_df[["LATITUDE", "LONGITUDE"]].dropna()
map_center = [42.2808, -83.7430]  # Ann Arbor, Michigan
weather_map = folium.Map(location=map_center, zoom_start=8)

stations_sample = stations.sample(n=min(100, len(stations)), random_state=42)  # Limit markers
for _, row in stations_sample.iterrows():
    folium.Marker(
        location=[row["LATITUDE"], row["LONGITUDE"]],
        icon=folium.Icon(color="blue", icon="cloud"),
    ).add_to(weather_map)

weather_map
```
